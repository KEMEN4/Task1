```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path


# =========================================================
# 1) WEATHER FILE PATH (adapt exact filename)
# =========================================================
in_path = r"C:\Users\User\Downloads\dataexport_20260216T123426.csv"

# =========================================================
# 2) ROBUST READING (skip metadata lines until "timestamp")
# =========================================================
raw = pd.read_csv(in_path)

start_idx = None
for i in range(len(raw)):
    if str(raw.iloc[i, 0]).lower() == "timestamp":
        start_idx = i + 1
        break
if start_idx is None:
    raise ValueError("Line 'timestamp' not found in CSV file.")

df = raw.iloc[start_idx:].copy()
df.columns = ["timestamp", "T_out_C"]
df["timestamp"] = pd.to_datetime(df["timestamp"], format="%Y%m%dT%H%M", errors="coerce")
df["T_out_C"] = pd.to_numeric(df["T_out_C"], errors="coerce")
df = df.dropna(subset=["timestamp", "T_out_C"]).sort_values("timestamp")
df = df.drop_duplicates(subset=["timestamp"], keep="first")
df = df.set_index("timestamp")
df.index.name = "timestamp"

# =========================================================
# 3) BUILDING PARAMETERS (article + assumptions)
# =========================================================
UA = 10348        # W/K  (article)
V  = 17988        # m3   (article)
ACH = 0.5         # 1/h  (your choice)
rho = 1.2         # kg/m3
Cp  = 1005        # J/kg/K

# Gains: q_g = q_sol + q_int  (if no solar, set q_sol=0)
occupants = 500
q_sol = 0.0
q_int = occupants * 100.0    # W (simple assumption)
q_g = q_sol + q_int

# Indoor temperatures (assumptions)
Tin_heat = 15.5  # °C
Tin_cool = 22    # °C

# =========================================================
# 4) ARTICLE FORMULAS: BLC and Tb
# BLC = UA + mCp with m_dot = rho*ACH*V/3600
# Tb = Tin - qg/BLC
# =========================================================
m_dot = rho * ACH * V / 3600.0   # kg/s
mCp   = m_dot * Cp               # W/K
BLC   = UA + mCp                 # W/K

Tb_heat = Tin_heat - q_g / BLC
Tb_cool = Tin_cool - q_g / BLC

print(f"BLC = {BLC:.2f} W/K")
print(f"Tb_heat = {Tb_heat:.2f} °C | Tb_cool = {Tb_cool:.2f} °C")


# =========================================================
# 5) HDD / CDD (with (+)) + q_heating/q_cooling loads
# HDD = sum (Tb - Tout)^+
# CDD = sum (Tout - Tb)^+
# q_heating = BLC*(Tb - Tout)^+
# q_cooling = BLC*(Tout - Tb)^+
# =========================================================
df["HDH_degC_h"] = np.maximum(Tb_heat - df["T_out_C"], 0.0)
df["CDH_degC_h"] = np.maximum(df["T_out_C"] - Tb_cool, 0.0)

# Hourly thermal loads (W) according to article
df["q_heating_W"] = BLC * df["HDH_degC_h"]
df["q_cooling_W"] = BLC * df["CDH_degC_h"]
df["q_heating_kW"] = df["q_heating_W"] / 1000.0
df["q_cooling_kW"] = df["q_cooling_W"] / 1000.0

# Daily aggregation (HDD/CDD in °C·day)
daily = df.resample("D").agg(
    Tmean_C=("T_out_C", "mean"),
    Tmin_C=("T_out_C", "min"),
    Tmax_C=("T_out_C", "max"),
    HDH_degC_h=("HDH_degC_h", "sum"),
    CDH_degC_h=("CDH_degC_h", "sum"),
    q_heating_kW=("q_heating_kW", "sum"),
    q_cooling_kW=("q_cooling_kW", "sum"),
)
daily["HDD_degC_day"] = daily["HDH_degC_h"] / 24.0
daily["CDD_degC_day"] = daily["CDH_degC_h"] / 24.0

# NH and NC (days with non-zero contribution)
daily["is_heating_day"] = daily["HDD_degC_day"] > 0
daily["is_cooling_day"] = daily["CDD_degC_day"] > 0
NH = int(daily["is_heating_day"].sum())
NC = int(daily["is_cooling_day"].sum())

HDD_total = float(daily["HDD_degC_day"].sum())
CDD_total = float(daily["CDD_degC_day"].sum())


# =========================================================
# 6) ANNUAL ENERGIES Qheating/Qcooling (kWh) — ARTICLE FORMULA
# Q(kWh) = BLC * HDD * 24 / 1000
# =========================================================
Q_heating_kWh = BLC * HDD_total * 24.0 / 1000.0
Q_cooling_kWh = BLC * CDD_total * 24.0 / 1000.0

# (Optional) verification via hourly integration
Q_heating_kWh_integral = df["q_heating_kW"].sum()
Q_cooling_kWh_integral = df["q_cooling_kW"].sum()

print("\n=== HDD/CDD ===")
print(f"NH = {NH} days | NC = {NC} days")
print(f"HDD total = {HDD_total:.2f} °C·day | CDD total = {CDD_total:.2f} °C·day")

print("\n=== Energies (kWh) ===")
print(f"Q_heating (article formula)  = {Q_heating_kWh:,.2f} kWh")
print(f"Q_heating (hourly integral)  = {Q_heating_kWh_integral:,.2f} kWh")
print(f"Q_cooling (article formula)  = {Q_cooling_kWh:,.2f} kWh")
print(f"Q_cooling (hourly integral)  = {Q_cooling_kWh_integral:,.2f} kWh")


# =========================================================
# 7) HP PERFORMANCE: COP(t) & EER(t) (Tables 2 & 3)
#    - COP: 35A for T<=3°C, 35B for T>3°C
#    - EER: Tw=6°C
# =========================================================
T = df["T_out_C"]

# COP 35A (Air temperature -15 to 3°C, water 35°C)
COP_35A = (
    5e-7*T**6 + 2e-5*T**5 + 0.0003*T**4
    + 0.0017*T**3 + 0.009*T**2 + 0.078*T + 2.5019
)

# COP 35B (Air temperature 3 to 20°C, water 35°C)
COP_35B = (
    -2e-7*T**6 + 2e-5*T**5 - 0.0004*T**4
    + 0.0066*T**3 - 0.0606*T**2 + 0.4206*T + 2.0482
)

df["COP_t"] = np.where(T <= 3.0, COP_35A, COP_35B)
df["COP_t"] = df["COP_t"].clip(lower=1.5, upper=8.0)

# EER Tw = 6°C
df["EER_t"] = (-4e-5*T**3 + 0.005*T**2 - 0.2806*T + 7.7483)
df["EER_t"] = df["EER_t"].clip(lower=1.5, upper=8.0)

# =========================================================
# 8) ELECTRICITY DEMAND (HP)
#    P_elec_heat = q_heat / COP(t)
#    P_elec_cool = q_cool / EER(t)
# =========================================================
df["P_heat_elec_kW"] = np.where(df["q_heating_kW"] > 0, df["q_heating_kW"] / df["COP_t"], 0.0)
df["P_cool_elec_kW"] = np.where(df["q_cooling_kW"] > 0, df["q_cooling_kW"] / df["EER_t"], 0.0)

df["P_HP_total_kW"] = df["P_heat_elec_kW"] + df["P_cool_elec_kW"]

E_HP_kWh = df["P_HP_total_kW"].sum()
print(f"Electricity (HP total) = {E_HP_kWh:,.2f} kWh")


# =========================================================
# 9) EXPORT CSV + SUMMARY
# =========================================================
out_csv = Path("Task1_hourly_profiles_Tout_heat_cool_COP_EER_elec.csv")
df.reset_index().to_csv(out_csv, index=False)

summary = pd.DataFrame({
    "metric": [
        "UA_W_per_K","V_m3","ACH_1_per_h","m_dot_kg_s","mCp_W_per_K","BLC_W_per_K",
        "occupants","q_g_W","Tin_heat_C","Tin_cool_C","Tb_heat_C","Tb_cool_C",
        "E_HP_total_kWh","start_timestamp","end_timestamp","n_hours"
    ],
    "value": [
        UA,V,ACH,m_dot,mCp,BLC,
        occupants,q_g,Tin_heat,Tin_cool,Tb_heat,Tb_cool,
        float(E_HP_kWh), str(df.index.min()), str(df.index.max()), int(len(df))
    ]
})

summary_path = Path("Task1_summary_HP.csv")
summary.to_csv(summary_path, index=False)

print("Saved CSV:", out_csv)
print("Saved summary:", summary_path)


# =========================================================
# 10) PLOTS (profiles) - daily mean for readability
# =========================================================
daily = df.resample("D").mean(numeric_only=True)

plt.figure()
daily["T_out_C"].plot()
plt.xlabel("Date"); plt.ylabel("°C")
plt.title("Outdoor temperature (daily mean)")
plt.tight_layout()
plt.show()

plt.figure()
daily["q_heating_kW"].plot(label="Heating load (kW_th)")
daily["q_cooling_kW"].plot(label="Cooling load (kW_th)")
plt.xlabel("Date"); plt.ylabel("kW")
plt.title("Thermal loads (daily mean)")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure()
daily["COP_t"].plot(label="COP(t)")
daily["EER_t"].plot(label="EER(t)")
plt.xlabel("Date"); plt.ylabel("-")
plt.title("Heat pump performance vs climate (daily mean)")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure()
daily["P_HP_total_kW"].plot()
plt.xlabel("Date"); plt.ylabel("kW")
plt.title("Heat pump electricity demand (daily mean)")
plt.tight_layout()
plt.show()
```

```python
"""
TASK 2 — Forecasting model (Supervised Regression) based ONLY on the teacher’s regression lectures
-----------------------------------------------------------------------------------------------

Goal (Task 2 statement):
"Create a model which has the historical load profiles as inputs, weather & climate data as features,
and forecasts future load profiles."

How we answer it (in the exact spirit of the regression course):
1) We formulate a SUPERVISED REGRESSION problem:
      - Inputs/features  X  = weather + time features
      - Outputs/targets  y  = historical load profiles (heating / cooling)
   This matches the course: regression = predict a continuous variable, supervised = labeled data.

2) We train an OLS (Ordinary Least Squares) baseline model:
      - OLS minimizes RSS(β) = Σ (y_i - X_i β)^2  (course: least squares)

3) We also test Ridge regression (L2-regularized least squares):
      - Ridge minimizes RSS(β) + λ ||β||^2
      - Used in the course to reduce variance / stabilize coefficients (esp. with correlated predictors).

4) We validate using a TIME-BASED split (train early period, test late period).
   This is important for time series: we avoid “future leaking into past”.

5) “Forecast future load profiles”:
   We create simple climate trajectories by applying a temperature delta (+2°C, +4°C),
   recompute HDH/CDH consistently, then predict future heating/cooling profiles for all hours (8760).
   This directly demonstrates the forecasting capability under climate change.

Outputs:
- Saved trained models (joblib)
- Test predictions CSV
- Future scenario profiles CSVs (+2C, +4C)
- Plots: real vs predicted on test; full-year baseline vs future scenarios; metrics

NOTE:
- This stays within regression theory (OLS/Ridge) from the lectures.
- No neural networks, no LSTM, no “extra” methods.
"""

# =========================
# 0) Imports
# =========================
import numpy as np
import pandas as pd
from pathlib import Path
import joblib
import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.base import clone


# =========================
# 1) Load Task 1 results (historical profiles)
# =========================
# These are the historical load profiles generated from your physics-based model in Task 1.
# They contain:
#   - weather data (T_out, and degree-hours HDH/CDH)
#   - loads (q_heating_kW, q_cooling_kW)
#   - timestamp (hourly)
CSV_PATH = Path("Task1_hourly_profiles_Tout_heat_cool_COP_EER_elec.csv")
df = pd.read_csv(CSV_PATH)

# Ensure timestamps are datetime and sorted
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values("timestamp").reset_index(drop=True)

# Quick sanity check (optional prints)
print("Columns:", df.columns.tolist())
print("Rows:", len(df))
print(df.head(3))


# =========================
# 2) Build features X (weather + time) and targets y (loads)
# =========================
# In supervised regression:
#   X = input features
#   y = labeled outputs (historical load profiles)
#
# We add time features because demand depends on hour/day/month patterns.
df["hour"] = df["timestamp"].dt.hour
df["dow"] = df["timestamp"].dt.dayofweek
df["month"] = df["timestamp"].dt.month
df["doy"] = df["timestamp"].dt.dayofyear

# Cyclic encoding (course idea: transform inputs to represent periodicity)
# Hour repeats every 24 hours; day-of-year repeats every 365 days.
df["hour_sin"] = np.sin(2*np.pi*df["hour"]/24.0)
df["hour_cos"] = np.cos(2*np.pi*df["hour"]/24.0)
df["doy_sin"]  = np.sin(2*np.pi*df["doy"]/365.0)
df["doy_cos"]  = np.cos(2*np.pi*df["doy"]/365.0)

# Features used (X):
# Weather features:
#   - T_out_C
#   - HDH_degC_h (heating degree-hours)
#   - CDH_degC_h (cooling degree-hours)
# Time features:
#   - hour_sin, hour_cos, doy_sin, doy_cos
# Categorical time:
#   - day of week, month (encoded by OneHot)
num_cols = ["T_out_C", "HDH_degC_h", "CDH_degC_h", "hour_sin", "hour_cos", "doy_sin", "doy_cos"]
cat_cols = ["dow", "month"]

X = df[num_cols + cat_cols]

# Targets y (historical loads):
y_heat = df["q_heating_kW"].astype(float)
y_cool = df["q_cooling_kW"].astype(float)


# =========================
# 3) Time-based train/test split (avoid leakage)
# =========================
# We keep the last 20% of the year as TEST (future-like).
# This matches good practice for time series forecasting evaluation.
split_idx = int(len(df) * 0.8)

X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_heat_train, y_heat_test = y_heat.iloc[:split_idx], y_heat.iloc[split_idx:]
y_cool_train, y_cool_test = y_cool.iloc[:split_idx], y_cool.iloc[split_idx:]


# =========================
# 4) Preprocessing: scaling + one-hot encoding
# =========================
# Course note: Ridge regression is sensitive to feature scaling because penalty uses β^2.
# So we standardize numeric features (mean=0, std=1).
# For categorical features (dow, month) we use one-hot encoding.
preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
    ],
    remainder="drop"
)


# =========================
# 5) Metrics function (for report)
# =========================
def compute_metrics(y_true, y_pred):
    """
    Metrics commonly used in forecasting:
    - MAE: mean absolute error
    - RMSE: root mean squared error
    - peak comparisons (important for sizing)
    """
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    peak_true = float(np.max(y_true))
    peak_pred = float(np.max(y_pred))
    return {
        "MAE": mae,
        "RMSE": rmse,
        "Peak_true_kW": peak_true,
        "Peak_pred_kW": peak_pred,
        "Peak_abs_err_kW": abs(peak_true - peak_pred),
    }


# =========================
# 6) OLS baseline model (Ordinary Least Squares)
# =========================
# OLS minimizes RSS(β) = Σ (y - Xβ)^2 (course: least squares regression).
ols_pipe = Pipeline([
    ("preprocess", preprocess),
    ("model", LinearRegression())
])

# IMPORTANT: train two separate models (heating and cooling)
# We must CLONE the pipeline, otherwise a second .fit() overwrites the first.
ols_heat = clone(ols_pipe).fit(X_train, y_heat_train)
ols_cool = clone(ols_pipe).fit(X_train, y_cool_train)

pred_ols_heat = ols_heat.predict(X_test)
pred_ols_cool = ols_cool.predict(X_test)

print("\n=== OLS (Least Squares) ===")
print("Heating:", compute_metrics(y_heat_test, pred_ols_heat))
print("Cooling:", compute_metrics(y_cool_test, pred_ols_cool))


# =========================
# 7) Ridge regression (L2 regularization)
# =========================
# Ridge minimizes: RSS(β) + λ ||β||^2 (course: shrinkage/regularization).
# We choose λ by time-series cross-validation (TimeSeriesSplit),
# which respects temporal order (no random mixing).
ridge_pipe = Pipeline([
    ("preprocess", preprocess),
    ("model", Ridge())
])

param_grid = {"model__alpha": np.logspace(-4, 4, 25)}  # alpha = λ
tscv = TimeSeriesSplit(n_splits=5)

ridge_cv_heat = GridSearchCV(
    clone(ridge_pipe), param_grid, cv=tscv,
    scoring="neg_mean_absolute_error", n_jobs=-1
)
ridge_cv_cool = GridSearchCV(
    clone(ridge_pipe), param_grid, cv=tscv,
    scoring="neg_mean_absolute_error", n_jobs=-1
)

ridge_cv_heat.fit(X_train, y_heat_train)
ridge_cv_cool.fit(X_train, y_cool_train)

best_ridge_heat = ridge_cv_heat.best_estimator_
best_ridge_cool = ridge_cv_cool.best_estimator_

pred_ridge_heat = best_ridge_heat.predict(X_test)
pred_ridge_cool = best_ridge_cool.predict(X_test)

print("\n=== Ridge (CV) ===")
print("Best alpha heat:", ridge_cv_heat.best_params_)
print("Heating:", compute_metrics(y_heat_test, pred_ridge_heat))
print("Best alpha cool:", ridge_cv_cool.best_params_)
print("Cooling:", compute_metrics(y_cool_test, pred_ridge_cool))


# =========================
# 8) Select final model (OLS vs Ridge) using test MAE
# =========================
# Course concept: choose model based on predictive performance (generalization).
mae_ols_heat = mean_absolute_error(y_heat_test, pred_ols_heat)
mae_ridge_heat = mean_absolute_error(y_heat_test, pred_ridge_heat)
final_heat = best_ridge_heat if mae_ridge_heat <= mae_ols_heat else ols_heat
final_heat_name = "Ridge" if mae_ridge_heat <= mae_ols_heat else "OLS"

mae_ols_cool = mean_absolute_error(y_cool_test, pred_ols_cool)
mae_ridge_cool = mean_absolute_error(y_cool_test, pred_ridge_cool)
final_cool = best_ridge_cool if mae_ridge_cool <= mae_ols_cool else ols_cool
final_cool_name = "Ridge" if mae_ridge_cool <= mae_ols_cool else "OLS"

print("\n=== Final choice ===")
print("Final heating model:", final_heat_name)
print("Final cooling model:", final_cool_name)


# =========================
# 9) Plot: Real vs Predicted on TEST period
# =========================
# These plots demonstrate that the learned regression function f(X) generalizes.
plt.figure(figsize=(12,4))
plt.plot(y_heat_test.values, label="Real heating (test)", color="blue", linewidth=3)
plt.plot(final_heat.predict(X_test), label=f"Pred heating ({final_heat_name})", color="red", linewidth=1.5)
plt.title("Heating load — test period: real vs predicted")
plt.xlabel("Test time index (hours)")
plt.ylabel("kW")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(12,4))
plt.plot(y_cool_test.values, label="Real cooling (test)", color="blue", linewidth=3)
plt.plot(final_cool.predict(X_test), label=f"Pred cooling ({final_cool_name})", color="red", linewidth=1.5)
plt.title("Cooling load — test period: real vs predicted")
plt.xlabel("Test time index (hours)")
plt.ylabel("kW")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


# =========================
# 10) Save models + test predictions (deliverables)
# =========================
out_dir = Path("task2_outputs_teacher_course")
out_dir.mkdir(parents=True, exist_ok=True)

joblib.dump(final_heat, out_dir / f"model_heating_{final_heat_name}.joblib")
joblib.dump(final_cool, out_dir / f"model_cooling_{final_cool_name}.joblib")

test_pred_df = pd.DataFrame({
    "timestamp": df["timestamp"].iloc[split_idx:].values,
    "q_heating_kW_true": y_heat_test.values,
    "q_heating_kW_pred": final_heat.predict(X_test),
    "q_cooling_kW_true": y_cool_test.values,
    "q_cooling_kW_pred": final_cool.predict(X_test),
})
test_pred_df.to_csv(out_dir / "test_predictions.csv", index=False)

print("\nSaved models + test predictions to:", out_dir.resolve())


# =========================
# 11) "Forecast future load profiles" using simple climate deltas (+2°C, +4°C)
# =========================
# This is the key point of Task 2:
# once trained on historical data, we can predict future loads by feeding future weather/climate features.
#
# Here we implement a simple "delta" climate scenario:
#   T_out_future = T_out_historical + ΔT
#
# But we must recompute HDH/CDH consistently, otherwise features become incoherent.
# To keep consistency with your dataset, we estimate the base temperatures used for HDH and CDH.
# (If you already know the bases, you can directly set them.)
mask_h = df["HDH_degC_h"] > 0
mask_c = df["CDH_degC_h"] > 0

Tbase_heat = float(np.median(df.loc[mask_h, "HDH_degC_h"] + df.loc[mask_h, "T_out_C"])) if mask_h.any() else 18.0
Tbase_cool = float(np.median(df.loc[mask_c, "T_out_C"] - df.loc[mask_c, "CDH_degC_h"])) if mask_c.any() else 22.0

print("\nEstimated base temperatures (from your dataset consistency):")
print("Tbase_heat =", Tbase_heat)
print("Tbase_cool =", Tbase_cool)

def make_future_scenario(df_hist: pd.DataFrame, delta_T: float) -> pd.DataFrame:
    """
    Create a future scenario DataFrame:
    - add delta_T to outdoor temperature
    - recompute HDH/CDH using estimated base temperatures
    - rebuild the SAME time features used in training
    - predict future heating/cooling loads using the trained regression models
    """
    df_f = df_hist.copy()

    # Climate perturbation: warmer climate
    df_f["T_out_C"] = df_f["T_out_C"] + delta_T

    # Recompute degree-hours to remain physically consistent with modified temperature
    df_f["HDH_degC_h"] = np.maximum(0.0, Tbase_heat - df_f["T_out_C"])
    df_f["CDH_degC_h"] = np.maximum(0.0, df_f["T_out_C"] - Tbase_cool)

    # Rebuild time features (must match training exactly)
    df_f["hour"] = df_f["timestamp"].dt.hour
    df_f["dow"] = df_f["timestamp"].dt.dayofweek
    df_f["month"] = df_f["timestamp"].dt.month
    df_f["doy"] = df_f["timestamp"].dt.dayofyear

    df_f["hour_sin"] = np.sin(2*np.pi*df_f["hour"]/24.0)
    df_f["hour_cos"] = np.cos(2*np.pi*df_f["hour"]/24.0)
    df_f["doy_sin"]  = np.sin(2*np.pi*df_f["doy"]/365.0)
    df_f["doy_cos"]  = np.cos(2*np.pi*df_f["doy"]/365.0)

    # Build X_future with the same feature columns
    X_future = df_f[num_cols + cat_cols]

    # Forecast (predict) future loads
    df_f["q_heating_future_kW"] = final_heat.predict(X_future)
    df_f["q_cooling_future_kW"] = final_cool.predict(X_future)

    # Physical constraint: loads cannot be negative
    df_f["q_heating_future_kW"] = np.maximum(0.0, df_f["q_heating_future_kW"])
    df_f["q_cooling_future_kW"] = np.maximum(0.0, df_f["q_cooling_future_kW"])

    return df_f

def annual_stats(df_hist: pd.DataFrame, df_future: pd.DataFrame):
    """
    Compare annual energy (sum of hourly kW -> approx kWh if 1-hour steps) and peaks.
    Important for energy system sizing.
    """
    E_heat_hist = df_hist["q_heating_kW"].sum()
    E_heat_fut  = df_future["q_heating_future_kW"].sum()

    E_cool_hist = df_hist["q_cooling_kW"].sum()
    E_cool_fut  = df_future["q_cooling_future_kW"].sum()

    P_heat_hist = df_hist["q_heating_kW"].max()
    P_heat_fut  = df_future["q_heating_future_kW"].max()

    P_cool_hist = df_hist["q_cooling_kW"].max()
    P_cool_fut  = df_future["q_cooling_future_kW"].max()

    print("\n=== Annual energy (approx kWh if hourly) ===")
    print(f"Heating: hist {E_heat_hist:.2f} -> future {E_heat_fut:.2f} "
          f"({(E_heat_fut/E_heat_hist-1)*100:.2f}%)")
    print(f"Cooling: hist {E_cool_hist:.2f} -> future {E_cool_fut:.2f} "
          f"({(E_cool_fut/E_cool_hist-1)*100:.2f}%)")

    print("\n=== Peak loads (kW) ===")
    print(f"Heating peak: hist {P_heat_hist:.2f} -> future {P_heat_fut:.2f} "
          f"({(P_heat_fut/P_heat_hist-1)*100:.2f}%)")
    print(f"Cooling peak: hist {P_cool_hist:.2f} -> future {P_cool_fut:.2f} "
          f"({(P_cool_fut/P_cool_hist-1)*100:.2f}%)")


# Create +2°C and +4°C scenarios (simple climate trajectories)
df_plus2 = make_future_scenario(df, delta_T=2.0)
df_plus4 = make_future_scenario(df, delta_T=4.0)

print("\n--- Scenario +2°C ---")
annual_stats(df, df_plus2)

print("\n--- Scenario +4°C ---")
annual_stats(df, df_plus4)


# =========================
# 12) Plot: Full-year baseline vs future scenarios
# =========================
# These plots are the clearest evidence that we "forecast future load profiles".
plt.figure(figsize=(12,4))
plt.plot(df["q_heating_kW"].values, label="Heating hist", color="blue", linewidth=1)
plt.plot(df_plus2["q_heating_future_kW"].values, label="Heating +2°C", color="red", linewidth=2)
plt.plot(df_plus4["q_heating_future_kW"].values, label="Heating +4°C", color="green", linewidth=3)
plt.title("Heating — full year: historical vs future climate deltas")
plt.xlabel("Hour of year")
plt.ylabel("kW")
plt.legend()
plt.grid(True, alpha=0.2)
plt.show()

plt.figure(figsize=(12,4))
plt.plot(df["q_cooling_kW"].values, label="Cooling hist", color="blue", linewidth=1)
plt.plot(df_plus2["q_cooling_future_kW"].values, label="Cooling +2°C", color="red", linewidth=2)
plt.plot(df_plus4["q_cooling_future_kW"].values, label="Cooling +4°C", color="green", linewidth=3)
plt.title("Cooling — full year: historical vs future climate deltas")
plt.xlabel("Hour of year")
plt.ylabel("kW")
plt.legend()
plt.grid(True, alpha=0.2)
plt.show()


# =========================
# 13) Export future profiles (for next tasks: clustering, PEDesigner)
# =========================
# This produces CSVs with timestamp + predicted future loads.
df_plus2[["timestamp", "q_heating_future_kW", "q_cooling_future_kW"]].to_csv(
    out_dir / "future_profiles_plus2C.csv", index=False
)
df_plus4[["timestamp", "q_heating_future_kW", "q_cooling_future_kW"]].to_csv(
    out_dir / "future_profiles_plus4C.csv", index=False
)

print("\nSaved future profiles:")
print("-", (out_dir / "future_profiles_plus2C.csv").resolve())
print("-", (out_dir / "future_profiles_plus4C.csv").resolve())

print("\nDONE: Task 2 answered with course-based regression (OLS/Ridge) + future scenario forecasting.")
```
